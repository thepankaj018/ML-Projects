```python
df_final = pd.DataFrame()

for index,rows in df.iterrows():
  match_id = []
  balls = []
  runs = []
  batting = []
  bowling = []
  batsman = []
  bowler = []
  player_out = []
  city = []
  venue = []
  
  
  mId = rows['match_id']
  bowl = rows['bowling']
  
  bt = rows['batting']
  c = rows['city']
  v = rows['venue']
 
  
  for delivery in rows['innings'][0]['1st innings']['deliveries']:
    for key in delivery.keys():
      match_id.append(mId)
      balls.append(key)
      runs.append(delivery[key]['runs']['total'])
      batting.append(bt)
      bowling.append(bowl)
      city.append(c)
      venue.append(v)
      bowler.append(delivery[key]['bowler'])
      batsman.append(delivery[key]['bowler'])
      try:
        player_out.append(delivery[key]['wicket']['player_out'])
      except:
        player_out.append('0')  


  ball_df = pd.DataFrame({
      'match_id':match_id,
      'batting_team':batting,
      'bowling_team':bowling,
      'balls':balls,
      'batsman':batsman,
      'bowler':bowler,
      'runs':runs,
      'player_out':player_out,
      'city':city,
      'venue':venue})
  

  df_final = df_final.append(ball_df)
     
```

# Current score

```python
# applying cumsum method of pandas to get cimmulative score
df['current_score'] = df.groupby('match_id').cumsum()['runs']
```

```python
df.head()
```

# Balls left

```python
df['balls'].dtypes
```

```python
# Adding running over of the match
df['over'] = df['balls'].astype(str).str.split(".").apply(lambda x : x[0])
```

```python
# Adding balls of the current over
df['ball_no'] = df['balls'].astype(str).str.split(".").apply(lambda x : x[1])
```

```python
# adding number of bowl bowled
df['balls_bowled'] = (df['over'].astype(int) * 6) + df['ball_no'].astype(int)
```

```python
# adding balls left
df['balls_left'] = 120 - df['balls_bowled'].astype(int)
```

```python
# Due to some extra balls some value of balls might be negative so making them equal to 0
df['balls_left'] = df['balls_left'].apply(lambda x: 0 if x < 0 else x)
```

## Wickets left

```python
# creating a dummy features which will values as 1 if some player got out.
df['dummy'] = df['player_out'].apply(lambda x : 0 if x == '0' else 1)
```

```python
# Applying group on match_id and taking cumsum on dummmy
df['wickets_fallen'] = df.groupby('match_id').cumsum()['dummy']
```

```python
# creating feature wickets_left
df['wickets_left'] = 10 - df['wickets_fallen']
```

## Current run rate

```python
df['crr'] = round((df['runs'] * 6)/df['balls_bowled'],2)
```

## Runs Scored in last five overs

```python
groups = df.groupby('match_id')

match_ids = df['match_id'].unique()

last_five = []

for elem in match_ids:
  last_five.extend(groups.get_group(elem).rolling(window = 30).sum()['runs'].values)
```

```python
len(last_five)
```

```python
df['last_five'] = last_five
```
